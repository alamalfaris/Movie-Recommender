## Data Cleaning

```
df.isna().sum()
```

```
feature = ['genres', 'keywords', 'cast', 'director']
for i in feature:
  df[i] = df[i].fillna('')
```

- Kita cuma fillna 4 feature di atas karena 4 ini yang akan digunakan untuk modelling

```
df.isna().sum()
```

## Feature Engineering

```
def combo(x):
  return x['genres'] + '' + x['keywords'] + '' + x['cast'] + '' + x['director']

df['combo_feature'] = df.apply(combo, axis=1)
df.head(2)
```

```
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```
movie_matrix = cv.fit_transform(df['combo_feature'])
```

## Machine Learning Model

```
from sklearn.metrics.pairwise import cosine_similarity
```

```
cos_score = cosine_similarity(movie_matrix)
cos_score
```

## Try use the model

```
fav_movie = input('Input your favorite movie: ')
```

```
df[df['original_title'] == fav_movie].index
```

```
def index_by_movie(movie):
  return list(df[df['title'] == movie]['index'])[0]
```

```
def movie_by_index(ind):
  return list(df[df['index'] == ind]['title'])[0]
```

```
index_mov = index_by_movie(fav_movie)
index_mov
```

```
similar_movie = list(enumerate(cos_score[index_mov]))
similar_movie
```

```
sorted_movie = sorted(similar_movie, key=lambda x: x[1], reverse=True)[1:]
sorted_movie
```

```
print(f"Top 5 movie recommendation based on your favorite movie: ")
count = 0
for i in sorted_movie:
  print(movie_by_index(i[0]))
  count += 1
  if count == 5:
    break
```
